```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
```

```python
# Define parameters for node embeddings
theta_1 = Parameter('θ1')
theta_2 = Parameter('θ2')
theta_3 = Parameter('θ3')

# Create Quantum Circuit with 3 qubits 
circuit = QuantumCircuit(3)

# Encode node features using RY rotations
circuit.ry(theta_1, 0)
circuit.ry(theta_2, 1)
circuit.ry(theta_3, 2)
```

```python

# Entanglement to model message passing 
circuit.cz(0, 1)  # Edge between Node 1 and Node 2
circuit.cz(1, 2)  # Edge between Node 2 and Node 3

# Apply trainable transformations (graph convolutional updates)
circuit.rx(theta_1, 0)
circuit.rx(theta_2, 1)
circuit.rx(theta_3, 2)

# Draw the circuit
print(circuit.draw())
```
